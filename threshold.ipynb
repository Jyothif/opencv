```python
img = cv2.resize(img, (320,240))
```

```python
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
#applying threshold to gray image
ret, thrsh1 = cv2.threshold(gray_img,125,255,cv2.THRESH_BINARY)
ret, thrsh2 = cv2.threshold(gray_img,125,255,cv2.THRESH_BINARY_INV)
ret, thrsh3 = cv2.threshold(gray_img,125,255,cv2.THRESH_TRUNC)
ret, thrsh4 = cv2.threshold(gray_img,125,255,cv2.THRESH_TOZERO)
ret, thrsh5 = cv2.threshold(gray_img,125,255,cv2.THRESH_TOZERO_INV)
```

```python
#Displaying resultant images
cv2.imshow("original image",img)
cv2.imshow('binary threshold',thrsh1)
cv2.imshow('binary threshold inverted',thrsh2)
cv2.imshow('Truncated threshold',thrsh3)
cv2.imshow('set to 0',thrsh4)
cv2.imshow('set to 0 inverted',thrsh5)


cv2.waitKey(0)
cv2.destroyAllWindows()
```
