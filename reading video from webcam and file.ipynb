# Reading and writing videos using opencv


These are the main functions in OpenCV video I/O
1. cv2.VideoCapture - creates a video capture object, which would help stream or display the video.
2. cv2.VideoWriter - saves output to the directory
3. cv2.imshhow(),cv2.waitKey(), and get() method which is used to video datasuch as frames,height,fps etc

```python
import cv2
vid_cap = cv2.VideoCapture(r"C:\Users\ppyad\Downloads\Video\2.mp4")
```

`isOpened() method returns a boolean that indicates whether or not,otherwise will get an error`,

`this method is not valid to webcam`

`The get() method takes a single argument from an enumerated list of options`

```python
if(vid_cap.isOpened()==False):
    print("Error opening the video file")
    
else:
    #get frame rate information
    fps = vid_cap.get(5)
    print("Frame per second:",fps,'FPS')
```

```python
#Get frame count
frame_count = vid_cap.get(7)
print("Frame Count:",frame_count)
```

```python
while(vid_cap.isOpened()):
    ret,frame = vid_cap.read()
    if ret== True:
        cv2.imshow("Frame:",frame)
        key = cv2.waitKey(20)
        
        if key == ord('q'):
            break
        else:
            break
#release the video capture object
vid_cap.release()
cv2.destroyAllWindows()
```

# Reading a video from the webcam


- Retrieve the image frame height and width, using the get() method.
- Initialize a video capture object (as discussed in the previous sections), to read the video stream into memory, using any of the sources previously described.
- Create a video writer object.
- Use the video writer object to save the video stream to disk. 

```python
import cv2
vid_capt = cv2.VideoCapture(0,cv2.CAP_DSHOW)

frame_width = int(vid_capt.get(3))
frame_height = int(vid_capt.get(4))
frame_size=(frame_width,frame_height)
fps=5

# initalize video writer object

output = cv2.VideoWriter('video_file.mp4',
                        cv2.VideoWriter_fourcc("M","J","P","G"),5,frame_size
                        )

while(vid_capt.isOpened()):
    
    ret,frame = vid_capt.read()
    if ret == True:
            output.write(frame)
    else:
        print('Stream disconnected')
        break
        
vid_capt.release()
output.release()
 
```

**The VideoWriter() class takes the following arguments:** 

- filename: pathname for the output video file
- apiPreference:  API backends identifier
- fourcc: 4-character code of codec, used to compress the frames (fourcc)
- fps: Frame rate of the created video stream
- frame_size: Size of the video frames
- isColor: If not zero, the encoder will expect and encode color frames. Else it will work with grayscale frames (the flag is currently supported on Windows only).
